# **Function to find columns with missing values**

```python
def columns_with_missing_values(df):
    missing_columns = df.columns[df.isnull().any()].tolist()
    return missing_columns
```

```python
columns_with_missing_values(df)
```

# Function to categorize columns on the basis of data types
**A dictionary is retured for key as datatype and values as a list of columns**

```python
def categorize_columns_by_dtype(df):
    dtype_groups = df.columns.groupby(df.dtypes)
    return dtype_groups
```

```python
dtype_groups = categorize_columns_by_dtype(df)
```

```python
for dtype_group_column in dtype_groups:
  print("\n")
  print(f"{dtype_group_column} : {dtype_groups.get(dtype_group_column)}")
```

# **Finding duplicates columns**

```python
def remove_columns_with_duplicates(df):

    # columns with duplicate elements
    columns_with_duplicates = [col for col in df.columns if df[col].nunique() < len(df[col])]

    # Print the DataFrame before removing columns with duplicates
    print("Before removing columns with duplicates:")
    print(df.columns)

    # Removing the identified columns with duplicate elements
    df_cleaned = df.drop(columns=columns_with_duplicates)
    print("\nAfter removing columns with duplicates:")
    print(df_cleaned.columns)
```

```python
remove_columns_with_duplicates(df)
```

# **Box plots to viualize the outliers in the numeric columns**

```python
def plot_boxplots_in_batches(df, batch_size=5):
    # Select only numeric columns
    numeric_columns = df.select_dtypes(include=['number']).columns
    nice_plots = [26,1,6,18,20,27,58,59,56,60,61,62,63,64,65,66,69,70,75,79,82]
    # Split numeric columns into batches
    batch_size = 5
    for i in range(0, len(numeric_columns), batch_size):
        # Get the batch of columns
        batch_columns = numeric_columns[i:i + batch_size]

        # Plot the boxplot for the current batch
        plt.figure(figsize=(4,5))
        sns.boxplot(data=df[batch_columns])
        # plt.yscale('log')
        plt.title(f'Boxplot for Columns {i+1} to {min(i+batch_size, len(numeric_columns))}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        # plt.savefig(f'/content/box_plots_all{i+1} to {min(i+batch_size, len(numeric_columns))}.png')
        plt.show()
```

# **1 Charts for STATE column distribution**

**Plotting pie chart for STATE column**

```python
category_counts = df['STATE'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    category_counts.values,  # Values to plot
    labels=category_counts.index,  # Labels for each slice
    autopct='%1.1f%%',  # Display percentage on slices
    startangle=90,  # Start from 90 degrees
    colors=['skyblue', 'lightgreen', 'orange'],  # Custom colors (optional)
    # explode=(0.1, 0, 0)  # Optional: explode one slice for emphasis
)
plt.title('Distribution of STATE column')
# plt.savefig('/content/STATE_pie_chart.png')
plt.show()
```

**Bar plot for QUANTITY column**

**Pie chart for QUANTITY Column with first 4 and rest as others**

```python
quantity_groups = df['QUANTITY'].apply(lambda x: x if x in [1, 2, 3, 4] else 'others')

quantity_counts = quantity_groups.value_counts()

quantity_percentage = 100 * quantity_counts / quantity_counts.sum()

labels = [f'<{index} >:: {count} ({percent:.1f}%)' for index, count, percent in zip(quantity_counts.index, quantity_counts, quantity_percentage)]

plt.figure(figsize=(4,4))
plt.pie(quantity_counts, labels=labels, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Pie Chart of QUANTITY (1, 2, 3, 4, and Others)')
# plt.savefig('/content/quantity_pie_chart.png')
plt.show()
```
